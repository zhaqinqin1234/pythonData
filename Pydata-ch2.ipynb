```python
time_zones =[rec['tz'] for rec in records if 'tz' in rec]
```

```python
def get_counts(sequence):
    counts={}
    for x in sequence:
        if x in counts:
            counts[x] +=1
        else:
            counts[x]=1
    return counts
get_counts(time_zones)
```

```python
counts = get_counts(time_zones)
```

```python
len(time_zones)
```

```python
# how to get top 10 time zone
def top_counts(count_dict, n=10):
    value_key_pairs = [(count, tz) for tz, count in count_dict.items()]
    value_key_pairs.sort()
    return value_key_pairs[-n:]
```

```python
top_counts(counts)
```

```python
# or from standard library collections.Counter class find the answer
from collections import Counter
counts= Counter(time_zones)
counts.most_common(10)
```

### suppose you want to analyze timezone according window or non-windows user

```python
cframe = frame[frame.a.notnull()]
```

```python
operating_system = np.where(cframe['a'].str.contains('Windows'), 'Windows', 'Not Windows')
```

```python
operating_system[:5]
```

```python
## step2: group by time zone and operationg system
by_tz_os = cframe.groupby(['tz', operating_system])
```

```python
agg_counts = by_tz_os.size().unstack().fillna(0)
agg_counts[:10]
```

```python
indexer = agg_counts.sum(1).argsort()
indexer[:10]
```

```python
count_subset = agg_counts.take(indexer)[-10:]
count_subset
```

```python
count_subset.plot(kind='barh', stacked=True)
```

```python
normed_subset = count_subset.div(count_subset.sum(1), axis =0)
normed_subset.plot(kind='barh', stacked = True)
```

## MovieLens 1M dataset

```python
import pandas as pd
unames =['user_id', 'gender', 'age', 'occupation', 'zip']
```

```python
users = pd.read_table("pydata-book/datasets/movielens/users.dat", sep='::', header = None, names =unames)
```

```python
rnames =['user_id', 'movie_id', 'rating','timestamp']
ratings = pd.read_table('pydata-book/datasets/movielens/ratings.dat', sep='::', header=None, names=rnames)
```

```python
mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('pydata-book/datasets/movielens/movies.dat', sep="::", header=None, names=mnames)
```

```python
users.head(5)
```

```python
data = pd.merge(pd.merge(ratings, users), movies)
```

```python
data.info()
```

```python
mean_raings = data.pivot_table('rating', index='title', columns='gender', aggfunc='mean')
```

```python
mean_raings.head(5)
```

```python
ratings_by_title = data.groupby("title").size()
ratings_by_title.head()
```

```python
active_titles = ratings_by_title.index[ratings_by_title >=250]
```

```python
mean_ratings = mean_raings.loc[active_titles]
```

```python
top_female_ratings = mean_ratings.sort_values("F", ascending = False)
```
